Para usar las funciones del m√©todo de parametrizaci√≥n es necesario que el mapeo reciba dos par√°metros de entrada, por lo que definimos nuevas funciones que sean casos particulares de cada mapeo. Dentro del m√≥dulo `mapeos` se encuentra definida la funci√≥n para el mapeo est√°ndar y su inverso, que recibe un punto x,y y un par√°metro.

```julia
estandar1(x,y)=Estandar(x,y,0.3)
estandarI(x,y)=EstandarI(x,y,0.3)
estandar2(x,y)=Estandar(x,y,1.5)
estandar3(x,y)=Estandar(x,y,0.7)
estandar4(x,y)=Estandar(x,y,6.5)
estandar5(x,y)=EstandarI(x,y,1.5)
#Definimos un mapeo usando precisi√≥n extendida
k=parse(BigFloat,"0.5")
estandar5(x,y)=Estandar(x,y,k)
k=parse(BigFloat,"0.3")
estandar6(x,y)=Estandar(x,y,k)
```

Usando las funciones `Estable` e `Inestable` del m√≥dulo `ParametrizacionMetodo` se calculan las variedades, √©stas funciones reciben el mapeo, el orden del polinomio que debe ser un entero, el punto fijo alrededor del cual se calcula la variedad, en forma de arreglo, el intervalo del par√°metro representado mediante un arreglo de dos entradas y por √∫ltimo el paso con el que se recorre el intervalo del par√°metro.

Las funciones `Estable`, `Inestable` regresan un arreglo con dos polinomios que son la parametrizaci√≥n de las variedades (**Ws_1,Wu_1**) junto con otro arreglo de dos polinomios que son aquellos que usaremos para calcular el error (**Es_1,Eu_1**), adem√°s de los valores y vectores propios (**eigvec_s_1,eigval_s_1**).

```julia
Orden1 = 20
p_fijo_1 = [0.,0.] 
t_1 = [-10.,10.]
delta_t_1 = 0.125
Ws_1,Es_1,eigvec_s_1,eigval_s_1 = Estable(estandar1, Orden1, p_fijo_1, t_1, delta_t_1)
Wu_1,Eu_1,eigvec_u_1,eigval_u_1 = Inestable(estandar1, Orden1, p_fijo_1, t_1, delta_t_1)
```

Visualizamos la parametrizacion obtenida.

```julia

function IterarMap(f, x_ini, n)   #Definimos una funci√≥n para iterar el mapeo

    x = x_ini[1]                      #                                     #Damos condiciones iniciales
    y = x_ini[2]                          #

    iteradosMapX = [x]
    iteradosMapY = [y]               #Definimos dos listas que tendran los valores de cada par ordenado de theta y P, y agregamos las condiciones iniciales

    for i=0:n              #iniciamos un ciclo de iteraciones donde se calculan x_n, y_n y se agregan a lalista correspodiente
        F = f(x,y)

        push!(iteradosMapX,F[1])
        push!(iteradosMapY,F[2])

        x = F[1]
        y = F[2]

    end

    return iteradosMapX, iteradosMapY  #La funcion iterados regresa las listas que corresponden a la trayectoria del

end

function GraficarEstandarMap(f)
    n = 50
    s=2pi/14.
    i=0.
    for p_i=0:s:2pi
        for x_i =0:s:2pi
            i=i+1
            x_ini=[p_i,x_i]
            a,b = IterarMap(f,x_ini,n)
            p = scatter(a,b,marker=".",s=0.1,color="gray")
        end
    end

end

function Evaluar(Tiempo, paso, Pol, PuntoFijo, col)
    ValX = Float64[]
    ValY = Float64[]

    push!(ValX, PuntoFijo[1])
    push!(ValY, PuntoFijo[2])
    #@show(ValX)
    for t = Tiempo[1]:paso:Tiempo[2]

        x = Float64(Pol[1](t))+PuntoFijo[1]
        y = Float64(Pol[2](t))+PuntoFijo[2]

        push!(ValX,x)
        push!(ValY,y)

    end

    p = plot(ValX,ValY,linestyle = "-",marker=",",color=col)



    title("espacio fase")
    xlabel(L"$\theta$")
    ylabel(L"$p$")
    xlim(0.,2*pi)


end

function GraficarEst(Tiempo,paso,f,Pol,PuntoFijo,col)
    GraficarEstandarMap(f)
    Evaluar(Tiempo,paso,Pol,PuntoFijo,col)

end

```

```julia
GraficarEst([0.,-3.],-0.2,estandar1,Ws_1,[0.,0.],"orangered")
GraficarEst([0.,-3.],-0.2,estandar1,Wu_1,[0.,2pi],"royalblue")
```
