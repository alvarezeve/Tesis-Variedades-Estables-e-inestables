```julia
function Rotor(x,y,a,b)
   
    x_n = x+y
    y_n = y+a*x_n*(x_n-1.)*e^(-x_n)
    #=
    x_n = x+a*y
    y_n = y+x_n*(x_n -1)*e^(-x_n)
    =#

    return [x_n,y_n]
end
```

```julia
function evaluar(Tiempo, paso,A,B)
    ValX=Float64[]
    
    ValY=Float64[]

    for t = Tiempo[1]:paso:Tiempo[2]
        
        x = evaluate(A,t)
        y = evaluate(B,t)
        
        push!(ValX,x)
        push!(ValY,y)
        
    end
    return ValX,ValY
    
end
```

```julia
a=parse(BigFloat,"3.4")
t1=[-big(55.),big(50.)]
t2=[-big(55.),big(50.)]
Polinomio1F,E1F,Cx1,Cy1 = MetParametrización(Rotor,1., 20, [big(1.),big(0.)], a, big(1.), 1, t1, 0.125)
Polinomio2F,E2F,Cx2,Cy2 = MetParametrización(Rotor,1., 20, [big(1.),big(0.)], a, big(1.), 2, t2, 0.125)
V=evaluar(t1,0.025,Polinomio2F[1],Polinomio2F[2])
B=evaluar(t2,0.025,Polinomio1F[1],Polinomio1F[2])
```

```julia
figure(1)
#pygui(true)
plot(B[1],B[2],color="r",label="Estable")
plot(V[1],V[2],color="b",label="Inestable")
xlim(-1,1.2)
ylim(-1,1)
#title("Mapeo exponencial con a=$a ,BigFloat, orden 40")
grid("on")
legend(loc="upper right",fancybox="true" )
```
