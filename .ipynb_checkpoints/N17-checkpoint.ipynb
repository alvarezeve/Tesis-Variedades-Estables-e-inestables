```julia
function Henon(x,y,a,b)
   
    x_n = a-b*y-x^2
    y_n = x 

    return [x_n,y_n]
end
```

```julia
function evaluar(Tiempo, paso,A,B)
    ValX=Float64[]
    
    ValY=Float64[]

    for t = Tiempo[1]:paso:Tiempo[2]
        
        x = evaluate(A,t)
        y = evaluate(B,t)
        
        push!(ValX,x)
        push!(ValY,y)
        
    end
    return ValX,ValY
    
end
```

```julia
function Parametro(a)
    x1 = -1.+(1.+a)^.5
    x2 = -1.-(1.+a)^.5
    return x1,x2
end
```

## Polinomios de orden 10-22 BigFloat

```julia
#a=parse(BigFloat,"0.4")
a=0.4
t=[-1750.,1750.]
x1,x2=Parametro(a)
Polinomio1,E1,ValProp1,CxB1,CyB1 = MetParametrización(Henon,1., 10, [x2,x2], a, 1., 2,t, 0.125)
Polinomio2,E2,ValProp2,CxB2,CyB2 = MetParametrización(Henon,1., 22, [x2,x2], a, 1., 2,t, 0.125)
Polinomio3,E3,ValProp3,CxB3,CyB3 = MetParametrización(Henon,1., 24, [x2,x2], a, 1., 2,t, 0.125)
Polinomio4,E4,ValProp4,CxB4,CyB4 = MetParametrización(Henon,1., 26, [x2,x2], a, 1., 2,t, 0.125)
Polinomio5,E5,ValProp5,CxB5,CyB5 = MetParametrización(Henon,1., 52, [x2,x2], a, 1., 2,t, 0.125)
Polinomio6,E6,ValProp6,CxB6,CyB6 = MetParametrización(Henon,1., 62, [x2,x2], a, 1., 2,t, 0.125)
Polinomio7,E7,ValProp7,CxB7,CyB7 = MetParametrización(Henon,1., 72, [x2,x2], a, 1., 2,t, 0.125)
```

```julia
B1=evaluar(t,0.025,Polinomio1[1],Polinomio1[2])
B2=evaluar(t,0.025,Polinomio2[1],Polinomio2[2])
B3=evaluar(t,0.025,Polinomio3[1],Polinomio3[2])
B4=evaluar(t,0.025,Polinomio4[1],Polinomio4[2])
B5=evaluar(t,0.025,Polinomio5[1],Polinomio5[2])
B6=evaluar(t,0.025,Polinomio6[1],Polinomio6[2])
B7=evaluar(t,0.025,Polinomio7[1],Polinomio7[2])
figure(1)
plot(B1[1],B1[2],label="O(10)")
plot(B2[1],B2[2],label="O(22)")
plot(B3[1],B3[2],label="O(24)")
plot(B4[1],B4[2],label="O(26)")
#plot(B5[1],B5[2],label="O(18)")
#plot(B6[1],B6[2],label="O(20)")
#plot(B7[1],B7[2],label="O(22)")
xlabel("x")
ylabel("y")
title("Variedades inestables Mapeo de Henón k=0.4 con BigFloat")
grid("on")
xlim(-10,10)
ylim(-10,10)
legend(loc="upper right",fancybox="true" )
figure(2)
plot(E1[1],log10(E1[2]),marker= "None", label="O(10)")
plot(E2[1],log10(E2[2]),marker= "None", label="O(22)")
plot(E3[1],log10(E3[2]),marker= "None", label="O(24)")
plot(E4[1],log10(E4[2]),marker= "None", label="O(26)")
#plot(E5[1],log10(E5[2]),marker= "None", label="O(18)")
#plot(E6[1],log10(E6[2]),marker= "None", label="O(20)")
#plot(E7[1],log10(E7[2]),marker= "None", label="O(22)")
xlabel("Parámetro t")
ylabel(L"log10($|| E||_{\infty})$")
title("Error con BigFloat ")
grid("on")
legend(loc="upper right",fancybox="true" )
#=
figure(3)
converXB=[]
converYB=[]
for i in 1:length(CxB)
    push!(converXB,CxB[i][1][1])
    push!(converYB,CyB[i][1][1])
end
plot(converXB,marker=".",label="coeficientes x")
plot(converYB,marker="_",label="coeficientes y")
title("Convergencia")
xlabel("Orden")
ylabel(L"$a_{n+1}/a_{n}$")
legend(loc="upper right",fancybox="true" )
=#

```
