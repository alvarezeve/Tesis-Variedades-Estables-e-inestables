El método que se esta implementando es para sistemas hamiltonianos de dos dimensiones, eso quiere decir que toda la información del mismo se encuentra en una sola función llamada función hamiltoniana.  Entonces podemos decir que la función hamiltoniana determina la trayectoria del sistema en términos del momento y la posición, todo en términos de un parámetro t, que no necesariamente representa el tiempo. La gráfica de estas dos variables es a lo que llamamos el espacio fase . Una característica de las ecuaciones de hamilton, obenidas a partir de la hamiltoniana con las derivadas parciales, es que preservan el volumen (2-dmensional) del espacio fase. Otra característica muy importante, y tal vez la más básica en términos de la estructura del espacio fase es que son simplécticos. Es decir quee para dos trayectorias suficienemente cercanas, es decir infinitesimalmente cercas, preservan el volumen que definen en el espacio fase en términos de flujo. Es decir que el voluen del paralelogramo formado por los vectores que definen las trayectorias es constante en toda la dinámica del sistema. Ua generalización de esta declaración es el llamado  teorema integal de Poincaré-Cartan.Este teorema declara que la acción integral al rededor de una trayectoria tiene el mismo valor que una trayectoria cercana a la primera. 


Ya se trabajó con el método de parametrización para el mapeo estándar pero haciendo cada paso por separado. Es decir se hizo el método paso a paso para entender cómo se podía eso implementar. Ahora se quiere implementar el método de manera más general.En este caso nuestra función hamiltoniana es $$f_{k}(\theta,p)$$ que depende de los parámetros de la posición angular $\theta$ y del momento $x$. En este caso nuestra función hamiltoniana es $$f_{k}(\theta,p)$$ que depende de los parámetros de la posición $\theta$ y del momento $p$.  Como recordamos el mapeo estándar se define como :
$$f_{k}(\theta,p)= \left\{ \begin{array}{lcc}
             \theta +p &   (mod2\pi )\\
             \\ p+k sin(\theta) & (mod2\pi)
             \end{array}
   \right.$$

Este sistema representa un rotor pateado , es decir una barra con momento de inercia I y longitud l, que está sujeta en un extremo a un pivote sin fricción, el otro extremo de la barra esta sometido a impulsos periódicos .
   
   $$x es \theta$$


La manera en la que implementaremos este método es usando una paqueteria desarrollada para el lenguae Julia, llamada TaylorSeries, que como su nombre lo indica se trata de un paquete de series de Taylor. En este paquete podemos encontrar cmo generar estas series además de poder operarlas.

```julia
using TaylorSeries
#@code_warntype
```

Dado que estamos usando el mapeo estandar para implemetar el método necesitamos definirlo como función. 

```julia
# necesitamos también definir el mapeo standar 
"""EstandarMap(θ,p,k)
   
   Función que define el mapeo estándar.
  

   Argumentos:
   -θ : posición
   -p : momento
    -k : constante del mapeo
"""
function EstandarMap(θ,p,k)
    x_n = mod([ θ+p, p+k*sin(θ+p) ], 2*pi)
    return x_n
    
end
```

Defnimos una función llamada PolinimioTaylor1 que será la que cree los polinomios de grado n que representan las variedades. La forma en la que se generan estos polinomios es introduciendo dos arreglos que contienen los coeficientes hasta el orden n-1, genera dos nuevos arreglos auxiliares a los que se les agrega el término n-ésimo "temporal". Es decir agrega un último término que no es el definitivo si no que depende de las variables del mapeo. Una vez que los arreglos tienen el último término agregado se usan para generar dos polinomios tipo Taylor1.TaylorN, puesto que los últimos coeficientes de el polinomio son variables de tipo TaylorN.

```julia
"""PolinomioTaylor1(g,Coefθ,CoefP)
   
   Es una función cuyo objetivo es recibir dos listas con valores iniciales y crear dos polinomios de grado g.
   -g : grado del polinomio
   - Coefθ,CoefP :  Arreglo que contiene los coeficientes iniciales, es del tipo Array{TaylorSeries.TaylorN{Float64}}
   

    Esta función regresa dos arreglos que contienen elementos del tipo Taylor1.TaylorN, creados con las listas iniciales.

"""
function PolinomioTaylor1(g::Real,Coefθ,CoefP)
    #=
    Donde L es un arreglo que contiene los coeficientes del polinomio, g es el grado del polinomio, Coefθ y CoefP son
    arreglos que contienen los coeficientes que forman la variedad.
    
    Separamos el primer orden puesto que es diferente de los otros.
    Creamos θ,p como variables tipo TaylorN de orden 1, recordemos que aquí no es importante el orden pues sólo las
    estamos usando como variables dentro de un Taylor1 =#
    θ,p = set_variables(Float64, "θ p", order=1)
    #especificamos que LX,LP son de arreglos que guardarán coeficientes del polinomio, sólo son auxiliares en esta función
    Lθ = Array{TaylorSeries.TaylorN{Float64}}(1)
    LP = Array{TaylorSeries.TaylorN{Float64}}(1)
    #usamos un condicional para separar el caso 1 del resto
    if g == 1
        taylor_θ = Taylor1([θ], g)
        taylor_p = Taylor1([p], g)
    #en el caso en que g>1 entonces usamos las listas que van guardando los coeficientes
    else
        #como CX,CP están guardando los coeficientes pero necesitamos agregar el último término que será una variable 
        Lθ = push!(Coefθ, θ) 
        LP = push!(CoefP, p)
        
        taylor_θ = Taylor1(Lθ, g)
        taylor_p = Taylor1(LP, g)
        
        
    end
    
    return [taylor_θ,taylor_p]
end
#Esta función regresa tx,tp que son Taylor1.TaylorN
    
        
 
```

Si recordamos que el lado derecho de la ecuación cohomológica se ve como 
$$\Sigma_{n=0}a_{n} \lambda^{n} t^{n}$$
$$\Sigma_{n=0}b_{n} \lambda^{n} t^{n}$$

entonces debemos tener definida esta parte. La función Vec$\lambda$ recibe los arreglos que contienen a las variables que conforman los coeficientes de las variedades, los mismos que en el caso de la función PolinomioTaylor1. También recibe el grado del polinomio a calcular y el valor propio. Con esta información genera dos polinomios tipo Taylor1.TaylorN que forman parte de un vector, representando la parte de la composición del lador derecho. Tomando el valor propio va agregando un elemento a la lista que contien las potencias del mismo. Con esta lista se hace una multiplicación término a término con las listas de los coeficientes de las variables 

```julia
#=Esta función toma el arreglo que contiene las lambdas que se van calculando, los coeficientes de los polinomios
y el orden de los mismos, lo que hace es generar el lado derecho de la ecuación cohomológica, multiplicando a_n*λ^n
y generando un polinomio de gradno g con estos coeficientes
=#
"""
Vecλ(λarray,g,CX,CP)
Es una función que calcula la parte derecha de la ecuación comohológica, es decir la parte que involucra el valor propio.
Regresa un arreglo de tipo TaylorSeries.TaylorN{Float64}


Sus argumentos son:
-g      :  grado del polinomio.
-λarray :  Arreglo que contiene el valor propio y sus potencias, hasta el grado g.
-CX,CP  :  Los arreglos con los polinomios que se calculan en PolinomioTaylor1.

"""
function Vecλ(λarray,g,Coefθ,CoefP)
    push!(λarray,λarray[2]^g)
    xλt=Taylor1(λarray.*Coefθ,g)
    pλt=Taylor1(λarray.*CoefP,g)
    λvec=[xλt,pλt]
    return λvec,λarray
end
```

Ahora necesitamos usar los polinomios que vamos calculando o mejor dicho necesitamos empezar con el punto fijo y comenzar a calcular los polinomios término a término usando las funciones que ya definimos. La función Variedades recibe el orden que queremos calcular del polinomio, el punto fijo, es decir  donde la variedad va a ser calculada , el parámetro del mapeo $\epsilon$ y el tipo de variedad : estable o inestable. 




        Lo primero que hace es crear dos arreglos que contendrán las constantes de los polinomios "$Coef\theta$,CoefP" (coeficientes de X, coeficietes de P), los valores iniciales de estos arreglos son las componentes del punto fijo: $Coef\theta=[Punto fijo],CoefP=[Punto fijo], \lambda =[Punto fijo]$. Tanto estos dos arreglos como el arreglo que contiene a los  valores propios deben de iniciar con este valor .


Posteriormente se utiliza un ciclo en el que se va recorriendo el orden del polinomo desde el primer coeficiente. Es decir cada que queramos calcular un polinomio de orden g comenza calculando desde el primero. Como el primer coeficiente es muy particular se separa de los otros casos, $g>1$ . Lo que hace es crear los polinomios usando la función PolinomioTaylor1 usando primero las listas que solo contienen al punto fijo $Coef\theta,CoefP,\lambda$. Ya creados los polinomios les aplica el mapeo estándar $EstandarMap(Polinomio de $\theta$, Polinomio de P, k)$ , si el orden que estamos calculando es el primero entonces le calcula el jacobiano al resultado y obtiene sus valores y vectores propios . Dependiendo de cuál variedad sea la que busquemos escoge el primer o segundo valor propio y por tanto el vector propio correspondiente. Estos valores serán los segundos integrantes de las listas que contienen los coeficientes $Coef\theta,CoefP,\lambda$. 
Después de calcular g=1  usa las listas actualizadas de $Coef\theta,CoefP,\lambda$  para nuevamente calcular con la función PolinomioTaylor1 los nuevos polinomios. Evalúa estos en la función del mapeo estándar para obtener los nuevos términos y entonces usa la ecuación cohomológica.
$$(f_{\epsilon} \circ P)(t)-(P \circ \lambda )(t) $$
Calcula el jacobiano al resultado de la resta anterior, recordemos que no es cero porque estamos apunto de empatar los valores de orden g nuevos. Busca si hay términos independientes en el sistema que resulta  y los ordena como vector, si los términos independientes son cero entonces significa que tenemos algo como :
 $$\mathbb{A}\textbf{x}=\textbf{0}$$
Por lo que si el determinante de $\mathbb{A}$ es distinto de cero la única solución para $\textbf{x}$ es la trivial. Sin embargo si el determinante no es cero y además los térmnos independientes so también diferentes de cero , entonces hay una solución para el sistema y lo que se hace es tomar la matriz $\mathbb{A}$ invertirla y multiplicarla por el vector independiente. 
$$\textbf{x}=\mathbb{A^{-1}}\textbf{v}$$

Así se encuentran las soluciones del sistema que representan los nuevos coeficientes de los polinomios, con ellas se actualizan las listas $Coef\theta,CoefP,\lambda$. Hace este procedimiento hasta que llega al orden deseado y devuelve los arreglos con los coeficientes que se calcularon. 

```julia
#Creamos una función que reciba el orden del polinomio , el punto fijo, el parámetro k y 
#el tipo de varidad que queremos(estable=1, inestable=2) además del mapeo 
"""
Variedades(orden, PuntoFijo,k,tipo_v)
Es una función que calcula las variedades del mapeo estándar. Usa las funciones de PolinomioTaylor1 y Vecλ para calcular los
polinomios de cada lado de la ecuación cohomológica y les aplica el mapeo dado. 



Argumentos:
-orden : se trata del orden del polinomio.
-PuntoFijo : ES el punto fijo donde queremos calcular la variedad.
-k     : Es la constante del mapeo.
-tipo_v : 1 si la variedad es estable, 2 si es inestable.


"""
function Variedades(orden, PuntoFijo,k,tipo_v)
   
    #definimos unas listas donde se guardarán los coeficientes  de todo el polinomio, tales deben ser
    # de tipo "Array{TaylorSeries.TaylorN{Int64},1}" dado que los términos que se van agregando 
    # en cada orden son de tipo TaylorN.
    Coefθ = [PuntoFijo[1] + TaylorN(0.)]
    CoefP = [PuntoFijo[2] + TaylorN(0.)]
    #@show(CX,CP)
    
    #λarray es la lista que contiene a los coeficientes del polinomio de λ
    λarray = [PuntoFijo[1]]
    #@show(λarray)


    for g in 1:orden
        #@show(g)
        #Calculamos el primer término del polinomio usando las funciones anteriores
        if g == 1
        #Llamamos a la función que calcula los polinomios de grado g
            t = PolinomioTaylor1(g,Coefθ,CoefP)
            #@show(typeof(t))
            #Aplicamos el mapeo a los polinomios que resultan de la función anterior.
            #Or1 = StandarMap(t[1],t[2],k)
            Or1 = EstandarMap(t[1],t[2],k)
            #@show(Or1)
            #Calculamos el jacobiano del Orden 1 para obtener sus valores y vectores propios.
            JPO = jacobian(Or1)
            #La matriz JacFl está deinida para obtener los valores de JPO como flotantes.          
            JacFl = Array{Float64}(2,2)
            for ind in eachindex(JPO)
                #@show(ind)
                JacFl[ind] = JPO[ind].coeffs[1]
            end
            eigval,eigvec = eig(JacFl)
            λ = eigval[tipo_v]
            #@show(λ)
            Coeficiente_θ,Coeficiente_P = eigvec[:,tipo_v]
            push!(Coefθ, Coeficiente_θ)
            push!(CoefP, Coeficiente_P)
            push!(λarray, λ)
            #@show(λarray)    
        else
            #@show(g)
            t = PolinomioTaylor1(g,Coefθ,CoefP)
            # aplicamos el mapeo estándar y al resultado le llamamos OrG por Orden g.
            OrG = EstandarMap(t[1],t[2],k)
            #@show(OrG)
            #@show(λarray[2])
            #agregamos el término correspondiente a λ 
            λ_vec,λarray=Vecλ(λarray,g,Coefθ,CoefP)
            λvec = [λ_vec[1],λ_vec[2]]
            
            #@show(λvec)
            
            # ahora ya tengo las dos partes de la ecuación y debo igualarlas para resolver.
            EcuaCohomo=OrG-λvec
            #@show(EcuaCohomo)
            # de esta ecuación necesitamos solo los de orden g, así que los extraemos manualmente 
            x_g=EcuaCohomo[1].coeffs[g+1]
            p_g=EcuaCohomo[2].coeffs[g+1]
            vec_orden_g=[x_g,p_g]
            JacOrdenG = jacobian(vec_orden_g)
            #calculamos el término independiene en la ecuación
            x_ind=EcuaCohomo[1].coeffs[g+1].coeffs[1].coeffs[1]
            p_ind=EcuaCohomo[2].coeffs[g+1].coeffs[1].coeffs[1]
            vec_ind=[-x_ind,-p_ind]
            #Con esta información podemos evaluar lo siguiente:
            # Si el vector de términos independientes es cero y el determinante del jacobiano es distinto de cero
            # entonces la solución a la ecuación cohomológica es la trivial
            if norm(vec_ind)==0.
                if det(JacOrdenG)!=0.
                    Coeficiente_θ,Coeficiente_P=[0.,0.]
                    Coefθ[g+1]=Coeficiente_θ
                    CoefP[g+1]=Coeficiente_P
                end
            else
                # Si el vector de términos independientes es distinto de ceroentonces necesitamos 
                #resolver la ecuación JacOrdenG[x_g,p_g]*[x,p]**=vec_ind[x_g,p_g]
                # entonces solo se trata de invertir el jacobiano y multiplicar con el vector del lado izquierdo
                TermG=JacOrdenG \ vec_ind
                Coeficiente_θ=TermG[1]
                Coeficiente_P=TermG[2]
                Coefθ[g+1]=Coeficiente_θ
                CoefP[g+1]=Coeficiente_P
            
            end
            
        end
    end
    return Coefθ,CoefP,λarray
end
            
        
```

Hasta este punto ya tenemos la información más valiosa, hemos calculado los coeficientes de los polinomios. Lo que haremos ahora es graficar esto en el espacio fase además de querer evaluar el error.

La función evaluar lo que hace es tomar dos polinomios tipo Taylor1.TaylorN y va sustituyendo el valor del parámetro que varía desde cero hasta un t máximo en pasos de por ejemplo 0.2 . Les calcula el módlo 2pi  a cada resultado de esta evaluación y los agrega a unas listas auxiliares que servirán para graficar. 

```julia
#esta es una función para evaluar el los polinomios que calculamos
# recibe como entrada el tiempo que es el valor del parámetro t, el paso
# que define en pasos de cuanto queremos ir evaluando y claro los polinomios A,B
"""
evaluar(Tiempo, paso,A,B)
Es una fución que recibe dos polinomios y los evalúa en términos del parámetro Tiempo. Regresa dos arreglos con 
los valores que resultaron de la evaluación en cada paso. 



Argumentos:
-Tiempo : Valor  hasta el cual se desea evaluar el parámetro t en los polinomios.
-paso   : Es el incremento que se le dará al parámetro para ir evaluando .
-A,B    : Son los polinomios que se desean evaluar.
"""
function evaluar(Tiempo, paso,A,B)
    ValX = Float64[]
    ValY = Float64[]

    for t = 0:paso:Tiempo
        x = mod2pi(evaluate(A,t))
        y = mod2pi(evaluate(B,t))
        push!(ValX,x)
        push!(ValY,y)
        
    end
    p = plot(ValX,ValY,color="red",linestyle="-",marker=",")
    
end
```

Para poder graficar el espaco fase del mapeo estándar necesitamos hacer una evaluación o barrido de valores en la función,para todos los valores de x,p en un intervalo. La función iterarMap crea las listas con los valores que resultan de este barrido

```julia
function iterarMap(f,x_i,p_i,n, k)   #Definimos una función para iterar el mapeo
    
    x = x_i                      #    
                                     #Damos condiciones iniciales
    y = p_i                          #

    iteradosMX = [x_i]

    iteradosMY = [p_i]               #Definimos dos listas que tendran los valores de cada par ordenado de theta y P, y agregamos las condiciones iniciales

    for i=0:n              #iniciamos un ciclo de iteraciones donde se calculan x_n, y_n y se agregan a lalista correspodiente            

        F = f(x_i,p_i, k)
        
        push!(iteradosMX,F[1])
        
        push!(iteradosMY,F[2])
            

        x_i = F[1]

        p_i = F[2]
    end

    return iteradosMX, iteradosMY  #La funcion iterados regresa las listas que corresponden a la trayectoria del 
end
```

```julia
#esta función llama la función que itera el mapeo para crear su gráfica.
function graficarMap(k)
    n = 100
    s=2pi/18

    for p_i=0:s:2pi
        for x_i =0:s:2pi
            t,p = iterarMap(EstandarMap,x_i,p_i,n, k)
            #p = scatter(p,t,marker=".",s=0.1)
            p = scatter(t,p,marker=".",s=0.1)
        end
    end
    
end
```

```julia
#esta función solo llama a las anteriores para que todo sea más compacto, y grafica tamto el mapeo como la variedad
function Graficar(Tiempo, paso, k,A,B)
    graficarMap(k)
    evaluar(Tiempo,paso,A,B)
end
```

Necesitamos también definir funciones para poder observar el error. Para ello es necesario definir la ecuación cohomológica, esta ecuación escrita en forma de resta nos dirá que tan creíble son nuestros cálculos. Al hacer la sustitución de los polinomios calculados en :
$$f_{\epsilon} \circ P - P\circ \lambda= R$$
Encontraremos que R es una fucnión distinta de cero. Ya que nuestro polinomio es aproximado y por tanto R nos ayudará a calcular el error.

La función PolinomioCohomo lo que hace es sustitutir los polinomios en la ecuación cohomológica, y regresa la función R.

```julia

"""
PolinomioCohomo(Pol_vec,λvec, k) 
Es una función que toma los polinomios y los sustitute en la ecuación cohomológica, con el fin de obtener la resta de las 
dos composiciones que aparecen en la misma, es decir evaluar una diferencia.




Argumentos:
-Pol_vec :  se trata de un arreglo que contiene los polinomios correspondientes a P
-λvec: Es la parte derecha de la ecuación de cohomológia.
-k : Es el parámetro del mapeo estándar.


"""
function PolinomioCohomo(Pol_vec,λvec, k)
    Map_vec=EstandarMap(Pol_vec[1],Pol_vec[2],k)
    Ec_Cohomo = Map_vec-λvec
    return Ec_Cohomo
end
```

Para evaluar el error no basta con ver la diferencia de las dos composiciones, hay que analizar que pasa para los valores en los que se está trabajando. Por ello necesitamos de una función error que dado el resultado de la función PolinomioCohomo lo evalúa para los valores de t en el intervalo que se usa. Es decir el resultado de la ecuación de cohomología es un polinomio que podemos evaluar para los mismos valores en los que evaluamos los polinomios calculados con el méodo para acercarnos a la variedad. Por ello es posible evaluar el error. La manera enla que se define es:
    $$|f_{\epsilon} \circ P- P\circ \lambda|=E$$
    
Donde la norma es la nosma infinito. 

```julia
"""
Err(Ec_Cohomo,Tiempo,paso) 
Toma el resultado de aplicar la función Ec_Cohomo, que es un arreglo de dos elementos, el valor máximo del parámetro t 
y los pasos en lo que se evalúa. 
"""function Err(Ec_Cohomo,Tiempo,paso)
    Val=Float64[]
    Tiem=Float64[]
    E1=Ec_Cohomo[1]
    E2=Ec_Cohomo[2]
    for t = 0:paso:Tiempo
        x = evaluate(E1,t)
        y = evaluate(E2,t)
        E=[x,y]
        
        norma = norm(E,Inf)
        push!(Val,norma)
        push!(Tiem,t)
    
    end
    return Tiem,Val
end
```

La función ErrorE calcula el error llamando a las funciones PolinomioCohomo y Err, recibe los arreglos de los polinomios y primero los evalúa en PolinomioCohomo, después al resultado de esto le aplica la función Err. Al finar regresa los valores del parámetro t así como del error en dos listas. 

```julia
function ErrorE(Pol_vec,λvec,k,T,paso)
    EcuaCohomo = PolinomioCohomo(Pol_vec,λvec,k)
    Tiempo,valor=Err(EcuaCohomo,T,paso)
    return Tiempo, valor
end
```

```julia
"""
    CalculoVariedad(k,Tiempo,paso,Pol_vec,λvec)

La función grafica la variedad en el espacio fase junto con el mapeo estándar, además regresa los valores del error cuando 
el parámetro t varía.  Esta función regresa los polinomios y el valor del error. 




Argumentos:
-k :  constante del mapeo.
-Tiempo :  valor máximo del parámetro en los polinomios. 
-paso : la cantidad que va aumentando el parámetro en cada evaluación.
-Pol_vec : arreglo de dos elementos que contiene a los polinomios de x,θ
λvec : arreglo de dos elementos que contiene el lado derecho de la ecuación cohomológica.
"""
function CalculoVariedad(k,Tiempo,paso,Pol_vec,λvec)
    B=Pol_vec[2]
    A=Pol_vec[1]
    #Graficar(Tiempo,paso,k,B,A)
    Graficar(Tiempo,paso,k,A,B)
    x1,y1=ErrorE(Pol_vec,λvec, k,Tiempo,paso)
    

    return A,B,x1,y1
end
    
```

El propósito es implementar esto de manera eficiente por lo que lo más apropiado es tener todo el proceso anterior en una
  sola función. Para ello creamos la función MetParametrización que reune todas las funciones anteriormente definidas para hacer el cálculo de las variedades mucho más simple. Esta función primero llama a la función Variedades , recordemos que esta a su vez calculaba las variedades dado un punto fijo y el orden del polinomio. Es decir en esta función está la parte más importante del método. Despues de tener ya los arreglos con los coeficientes de los polinomios lo que queremos hacer es tomar sólo los valores de forma Float64, esto ya que los coeficientes que resultan de el cálculo de la variedad son de tipo TaylorN, por lo que extraemos sólo el valor del coeficiente como un Float64 y los usamos para crear polinomios que solo sean Taylor1.TaylorN . De igual manera creamos los de $ \lambda$. Por último graficamos y calculamos el error con la función CalculoVariedad.

```julia
function MetParametrización(orden,PuntoFijo,k,tipo_v,Tiempo,paso)
    CoeficientesX,CoeficientesP,λarray = Variedades(orden,PuntoFijo,k,tipo_v)
    X = Float64[]
    P = Float64[]
    for i in 1:orden+1
        push!(X,CoeficientesX[i].coeffs[1].coeffs[1])
        push!(P,CoeficientesP[i].coeffs[1].coeffs[1])
    end
    TaylorX = Taylor1(X, orden) 
    TaylorP = Taylor1(P, orden)
    Taylor = [-TaylorX,-TaylorP]
    
    λx_pol=Taylor1(X.*λarray,orden)
    λp_pol=Taylor1(P.*λarray,orden)
    λ_vec=[-λx_pol,-λp_pol]
    
    VariedadX, VariedadP,ErrorT,ErrorX=CalculoVariedad(k,Tiempo,paso,Taylor,λ_vec)
    return TaylorX,TaylorP,ErrorT, ErrorX
end
    
    
```

Queremos ahora observar como se modifica la variedad inestable para dferentes ordenes de polinomios. Comenzaremos con un orden 5 hasta terminar en un orden al rededor de 60. Conviene que el orden del polinomio sea impar dado que en este caso los órdenes pares de los polinomios son cero.
