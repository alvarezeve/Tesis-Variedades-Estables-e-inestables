```julia
function evaluar(A,Tiempo,paso,PuntoFijo,col)
    ValX=Float64[]
    ValY=Float64[]
    #push!(ValX, PuntoFijo[1]) 
    #push!(ValY, PuntoFijo[2])
    
    for t = Tiempo[1]:paso:Tiempo[2]
        x = A[1](t)
        y = A[2](t)
        
        push!(ValX,x)
        push!(ValY,y)
        
    end
    p = plot(ValX,ValY,linestyle="-",marker =",",color=col)
    title("espacio fase")
    xlabel(L"x")
    ylabel(L"y")
    return ValX,ValY
    
end
```

```julia
function parametro(a)
    x1 = -1. + (1. + a)^.5
    x2 = -1. - (1. + a)^.5
    return [x2,x2]
end
    
```

# Encontrando puntos homoclínicos

Para encontrar puntos homoclinicos definimos la funcion que genera el polinomio de la resta de las variedades

```julia
function Interseccion(p,U,S)
    (x,y)=p
    V= SVector( U[1](x)-S[1](y), U[2](x)-S[2](y) )
    
end
```

Definimos un intervalo donde se buscara la raiz.

```julia
Inter4(p)=Interseccion(p,Wu1,Ws1)
```

Esos puntos serán nuestro nuevo punto de partida para la parametrización.

Ya que sabemos cual es la primera interseción trasladamos nuestro sistema a ese punto... que en principio no es lo mismo que trasladar la parametrización ni tampoco lo mismo que comenzar a parametrizar desde ese punto... o si?

```julia
"""
HenonT(x::Any,y::Any,a::Real,b::Real,x0::Any,y0::Any)

EL mapeo de Hénon con dos parámetros trasladado a x0,y0.
"""
function HenonT(x::Any, y::Any, a::Real, b::Real, x0::Any, y0::Any)
    x_n = a-b*(y-y0)-(x-x0)^2
    y_n = x-x0
    return [x_n,y_n]
end
```

```julia
"""
HenonIT(x::Any, y::Any,a::Real,b::Real,x0::Any,y0::Any)
Función que implementa el mapeo de Hénon inverso trasladado
"""
function HenonIT(x::Any, y::Any,a::Real, b::Real, x0::Any, y0::Any)
    x_n = y-y0
    y_n = (x-x0+(y-y0)^2-a)/(-b)
    return [x_n,y_n]
end
```

```julia
a=1.5
henon1T(x,y)=HenonT(x,y,a,1.,pto_fijo_nuevo[1],pto_fijo_nuevo[2])
henonI1T(x,y)=HenonIT(x,y,a,1.,pto_fijo_nuevo[1],pto_fijo_nuevo[2])
punto1T = parametroT(a,pto_fijo_nuevo[1],pto_fijo_nuevo[2])

@show("pasa1")

Orden = 105
tiempo = [0.,3.5]
delta_t = 0.125
Wu1T,fu1T,vecu1T,valu1T = Inestable(henon1T,Orden ,punto1T,tiempo, delta_t);
Ws1T,fs1T,vecs1T,vals1T = Inestable(henonI1T, Orden , punto1T, tiempo, delta_t);


pygui(false)
figure(1)
evaluar(Wu1T, [0.,30.], 0.025, punto1, "royalblue")
evaluar(Ws1T, [0.,-30], -0.025, punto1, "orangered")
scatter([punto1T[1]],[punto1T[2]],color="lawngreen")
xlim(-10,10)
ylim(-10,10)
```
