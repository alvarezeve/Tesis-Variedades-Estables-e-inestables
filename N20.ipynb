```julia
function Rotor(x,y,a)
   
    x_n = x+y
    y_n = y+a*x_n*(x_n-1.)*e^(-x_n)

    return [x_n,y_n]
end
```

```julia
function evaluar(Tiempo, paso,A,B)
    ValX=Float64[]
    
    ValY=Float64[]

    for t = Tiempo[1]:paso:Tiempo[2]
        
        x = mod(evaluate(A,t),1)
        y = mod(evaluate(B,t),1)
        
        push!(ValX,x)
        push!(ValY,y)
        
    end
    return ValX,ValY
    
end
```

```julia
function Arnold(x,y)
    x_n = mod(2*x+y,1)
    y_n = mod(x+y,1)
    return [x_n,y_n]
end
```

```julia
ta=[-3.,3.]
PolArnold,ErrA=Inestable(Arnold,25,[0.,0.],ta,0.125)
AA=evaluar(ta,0.025,PolArnold[1],PolArnold[2])
```

```julia
function Henon(x,y,a,b)
   
    x_n = a-b*y-x^2
    y_n = x 

    return [x_n,y_n]
end
```
