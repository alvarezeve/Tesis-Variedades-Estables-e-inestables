```julia
function evaluar(Tiempo, paso,A,B)
    ValX=Float64[]
    
    ValY=Float64[]

    for t = Tiempo[1]:paso:Tiempo[2]
        
        x = mod(evaluate(A,t),1)
        y = mod(evaluate(B,t),1)
        
        push!(ValX,x)
        push!(ValY,y)
        
    end
    return ValX,ValY
    
end
```

```julia
function Arnold(x,y)
    x_n = mod(2*x+y,1)
    y_n = mod(x+y,1)
    return [x_n,y_n]
end
```
